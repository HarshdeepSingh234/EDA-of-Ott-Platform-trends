```python
# STEP 0: Imports & Load Data (INTERMEDIATE EDA PIPELINE)
# --------------------------------------------------------
# This notebook walks you through an interview-ready, step-by-step EDA
# for your OTT TV shows dataset. It uses only pandas + matplotlib (as requested).

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime


# Load your dataset
PATH = "/content/cleaned_tvshows_data.csv"
df = pd.read_csv(PATH)

# Quick peek
basic_info = {
    "rows": len(df),
    "columns": list(df.columns),
    "null_counts": df.isna().sum().to_dict()
}
basic_info
```

```python
# STEP 1: Clean & Standardize
# ---------------------------
df_clean = df.copy()

# 1A) Standardize strings (strip spaces)
for col in ["type", "title", "director", "cast", "country", "rating", "listed_in"]:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].astype(str).str.strip()

# 1B) Handle missing country (fill with 'Unknown')
if "country" in df_clean.columns:
    df_clean["country"] = df_clean["country"].replace({"": np.nan})
    df_clean["country"] = df_clean["country"].fillna("Unknown")

# 1C) Convert date_added to datetime + derive month/year if needed
df_clean["date_added"] = pd.to_datetime(df_clean["date_added"], errors="coerce")
df_clean["added_year"] = df_clean["date_added"].dt.year
df_clean["added_month"] = df_clean["date_added"].dt.month

# 1D) Ensure numeric types
df_clean["release_year"] = pd.to_numeric(df_clean["release_year"], errors="coerce")
df_clean["year_added"] = pd.to_numeric(df_clean["year_added"], errors="coerce")
df_clean["no_of_seasons"] = pd.to_numeric(df_clean["no_of_seasons"], errors="coerce")

# 1E) Deduplicate on (title, release_year) as a safe key for shows
before = len(df_clean)
df_clean = df_clean.drop_duplicates(subset=["title", "release_year"])
after = len(df_clean)
dedup_removed = before - after

dedup_removed
```

```python
# STEP 2: Feature Engineering
# ---------------------------
CURRENT_YEAR = datetime.now().year

# 2A) Content Age & Recency Flag
df_clean["content_age"] = CURRENT_YEAR - df_clean["release_year"]
df_clean["is_recent"] = df_clean["release_year"].ge(CURRENT_YEAR - 2)  # last 2 years

# 2B) Genre split & explode (from 'listed_in')
def split_genres(s: str):
    if pd.isna(s):
        return []
    return [g.strip() for g in str(s).split(",") if g.strip()]

df_clean["genre_list"] = df_clean["listed_in"].apply(split_genres)
df_genre = df_clean.explode("genre_list").rename(columns={"genre_list": "genre"})

# Quick sanity previews
display("Cleaned Unique Shows (Preview)")
display(df_clean.head(10))
display("Genre-Exploded View (Preview)")
display(df_genre.head(10))

df_clean.shape, df_genre.shape
```

```python
# OPTIONAL CORRECTION:
# The raw file seems "exploded" by genre already (multiple rows per title).
# To rebuild a unique "show-level" table with *all* genres merged, let's aggregate from the ORIGINAL df.

def merge_unique_genres(series: pd.Series) -> str:
    # Join all values, split by comma, unique + sorted, then join back
    vals = ",".join(series.dropna().astype(str))
    parts = [p.strip() for p in vals.split(",") if p.strip()]
    unique_sorted = sorted(set(parts))
    return ", ".join(unique_sorted)

# Group at show-level
grp_cols = ["title", "release_year"]
agg_map = {
    "type": "first",
    "director": "first",
    "cast": "first",
    "country": "first",
    "date_added": "first",
    "rating": "first",
    "year_added": "first",
    "no_of_seasons": "first",
    "listed_in": merge_unique_genres
}

df_show = df.groupby(grp_cols, as_index=False).agg(agg_map)

# Re-apply cleaning from Step 1 quickly to df_show
df_show["date_added"] = pd.to_datetime(df_show["date_added"], errors="coerce")
df_show["year_added"] = pd.to_numeric(df_show["year_added"], errors="coerce")
df_show["no_of_seasons"] = pd.to_numeric(df_show["no_of_seasons"], errors="coerce")
df_show["country"] = df_show["country"].fillna("Unknown")

# Add engineered fields
df_show["content_age"] = datetime.now().year - df_show["release_year"]
df_show["is_recent"] = df_show["release_year"].ge(datetime.now().year - 2)

# Build genre-exploded view properly
df_show["genre_list"] = df_show["listed_in"].apply(lambda s: [g.strip() for g in str(s).split(",") if g.strip()])
df_show_genre = df_show.explode("genre_list").rename(columns={"genre_list": "genre"})

# Previews
display("Show-level Table (genres merged)", df_show.head(10))
display("Show-level (genre-exploded)", df_show_genre.head(15))

df_show.shape, df_show_genre.shape
```

```python
# STEP 3: Univariate EDA (Counts & Distributions)
# -----------------------------------------------

# 3A) Content Type distribution
type_counts = df_show["type"].value_counts().reset_index()
type_counts.columns = ["type", "count"]
display("Content Type Distribution", type_counts)

plt.figure()
plt.bar(type_counts["type"], type_counts["count"])
plt.title("Content Type Distribution")
plt.xlabel("Type")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 3B) Ratings distribution (top 15)
rating_counts = df_show["rating"].value_counts().head(15).reset_index()
rating_counts.columns = ["rating", "count"]
display("Ratings Distribution (Top 15)", rating_counts)

plt.figure()
plt.bar(rating_counts["rating"], rating_counts["count"])
plt.title("Ratings Distribution (Top 15)")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# 3C) Top Genres (overall)
top_genres = (
    df_show_genre["genre"].value_counts()
    .head(15)
    .reset_index()
    .rename(columns={"index": "genre", "genre": "count"})
)
# Fix the rename: after value_counts(), resulting columns are index & genre
top_genres.columns = ["genre", "count"]
display("Top Genres (Overall)", top_genres)

plt.figure()
plt.barh(top_genres["genre"][::-1], top_genres["count"][::-1])
plt.title("Top Genres (Overall)")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()

# 3D) Seasons distribution (TV Shows)
if "no_of_seasons" in df_show.columns:
    seasons_counts = df_show["no_of_seasons"].value_counts().sort_index().reset_index()
    seasons_counts.columns = ["no_of_seasons", "count"]
    display("Number of Seasons Distribution", seasons_counts)

    plt.figure()
    plt.bar(seasons_counts["no_of_seasons"].astype(int), seasons_counts["count"])
    plt.title("Number of Seasons Distribution")
    plt.xlabel("Seasons")
    plt.ylabel("Count of Shows")
    plt.tight_layout()
    plt.show()
```

```python
# STEP 4: Bivariate EDA (Relationships)
# -------------------------------------

# 4A) Genre popularity by Rating (Top 8 genres)
top8_genres = df_show_genre["genre"].value_counts().head(8).index.tolist()
g_by_rating = (
    df_show_genre[df_show_genre["genre"].isin(top8_genres)]
    .groupby(["rating", "genre"])["title"]
    .count()
    .reset_index(name="count")
    .sort_values(["genre", "count"], ascending=[True, False])
)

display("Genre Popularity by Rating (Top 8 Genres)", g_by_rating.head(50))

# Draw stacked bars: ratings distribution within each top genre
pivot_gr = g_by_rating.pivot(index="genre", columns="rating", values="count").fillna(0)

plt.figure()
# For stacked bars, we'll iterate columns
bottom = np.zeros(len(pivot_gr))
for col in pivot_gr.columns:
    plt.bar(pivot_gr.index, pivot_gr[col].values, bottom=bottom, label=str(col))
    bottom += pivot_gr[col].values
plt.title("Ratings Mix within Top Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.legend(loc="upper right", bbox_to_anchor=(1.15, 1.0))
plt.tight_layout()
plt.show()

# 4B) Seasons vs Rating: average seasons by rating
avg_seasons_by_rating = (
    df_show.dropna(subset=["no_of_seasons"])
    .groupby("rating")["no_of_seasons"]
    .mean()
    .round(2)
    .reset_index(name="avg_seasons")
    .sort_values("avg_seasons", ascending=False)
)
display("Average Number of Seasons by Rating", avg_seasons_by_rating)

plt.figure()
plt.bar(avg_seasons_by_rating["rating"], avg_seasons_by_rating["avg_seasons"])
plt.title("Average Number of Seasons by Rating")
plt.xlabel("Rating")
plt.ylabel("Avg Seasons")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# STEP 5: Time-Series Trends
# --------------------------

# 5A) Releases per Year (by release_year)
yearly_releases = (
    df_show.groupby("release_year")["title"]
    .count()
    .reset_index(name="count")
    .sort_values("release_year")
)
display("Releases per Year", yearly_releases.tail(15))

plt.figure()
plt.plot(yearly_releases["release_year"], yearly_releases["count"], marker="o")
plt.title("Releases per Year")
plt.xlabel("Release Year")
plt.ylabel("Count of Shows")
plt.tight_layout()
plt.show()

# 5B) Additions per Year (year_added) - where available
avail = df_show["year_added"].notna().sum()
if avail > 0:
    added_per_year = (
        df_show.dropna(subset=["year_added"])
        .groupby("year_added")["title"]
        .count()
        .reset_index(name="count")
        .sort_values("year_added")
    )
    display("Additions per Year (Based on year_added)", added_per_year.tail(15))

    plt.figure()
    plt.plot(added_per_year["year_added"], added_per_year["count"], marker="o")
    plt.title("Additions per Year (Catalog Adds)")
    plt.xlabel("Year Added")
    plt.ylabel("Count of Shows Added")
    plt.tight_layout()
    plt.show()
```

```python
# STEP 6: Geographic Insights
# ---------------------------

geo = df_show.copy()
geo["country"] = geo["country"].fillna("Unknown").astype(str)
geo["country_list"] = geo["country"].apply(lambda s: [c.strip() for c in s.split(",") if c.strip()])
geo = geo.explode("country_list").rename(columns={"country_list": "country_norm"})

top_countries = (
    geo["country_norm"].value_counts().head(15).reset_index()
    .rename(columns={"index": "country", "country_norm": "count"})
)
top_countries.columns = ["country", "count"]
display("Top Content-Producing Countries (TV Shows)", top_countries)

plt.figure()
plt.barh(top_countries["country"][::-1], top_countries["count"][::-1])
plt.title("Top Content-Producing Countries (TV Shows)")
plt.xlabel("Count of Shows")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
# STEP 7: Auto-Generated Insight Snippets (for interview talking points)
# ---------------------------------------------------------------------

insights = {}

# Top 5 genres
insights["top_5_genres"] = df_show_genre["genre"].value_counts().head(5).to_dict()

# Ratings mix (top 5)
insights["ratings_mix_top5"] = df_show["rating"].value_counts(normalize=True).head(5).mul(100).round(1).to_dict()

# Median number of seasons overall
if df_show["no_of_seasons"].notna().any():
    insights["median_seasons"] = float(df_show["no_of_seasons"].median())

# Recent share (last 2 years)
insights["recent_share_%"] = round(df_show["is_recent"].mean() * 100, 1)

# Top countries (top 5)
insights["top_countries_top5"] = geo["country_norm"].value_counts().head(5).to_dict()

insights
```
